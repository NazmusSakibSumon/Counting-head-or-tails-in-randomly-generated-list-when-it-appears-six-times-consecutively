```python
import random
 
if __name__ == '__main__':
    h = ('h')
    t = ('t')
    sam=[]
    for i in range(0,10001):
        rand = (h, t)[random.randint(0, 1)]
        
        sam.append(rand)
        
    #rint(sam)
    
from itertools import groupby
groups = groupby(sam)
result = [(label, sum(1 for _ in group)) for label, group in groups]
print(result)
```

```python
count1=0
for (x,y) in result:
    if y==6:
        #print(x,y)
        count1+=1
print('total head or tail six times is: '+str(count1))
```

```python
count2=0
for (x,y) in result:
    if y==6:
        if x==h:
            #print(x,y)
            count2+=1
print('total head six times is: '+str(count2))
```

```python
count3=0
for (x,y) in result:
    if y==6:
        if x==t:
            #print(x,y)
            count3+=1
print('total tail six times is: '+str(count3))
```
